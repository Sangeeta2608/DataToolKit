#2.How does broadcasting work in NumPy?
Broadcasting is one of the most powerful features in NumPy—it lets you perform arithmetic operations on arrays of different shapes without writing explicit loops or reshaping manually.
Broadcasting describes how NumPy automatically expands the dimensions of arrays so they can be compatible for element-wise operations.
If the shapes of two arrays are not the same, NumPy tries to "broadcast" them to make their shapes compatible by stretching the smaller array along the mismatched dimension without actually copying data.

#21.Why is NumPy's array slicing faster than Python's list slicing?
NumPy's array slicing is faster than Python's native list slicing comes down to a few key differences in memory layout, data types, and internal implementation.
NumPy arrays are stored in contiguous blocks of memory,whereas Python lists are arrays of pointers to objects, which can be scattered all over memory.
This means NumPy can access slices by simple pointer arithmetic.Python lists, on the other hand, require dereferencing each element individually.
NumPy arrays contain elements of the same data type, enabling highly optimized operations.Python lists can mix types, so slicing has to account for each element's type and overhead.
NumPy slicing doesn't duplicate memory unless explicitly told to.Python list slicing creates a new list every time, copying each element.

```python
#2.Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Score': [85, 90, 88, 92]
}
df = pd.DataFrame(data)
mean_age = df['Age'].mean()
mean_score = df['Score'].mean()
print("Mean Age:", mean_age)
print("Mean Score:", mean_score)
```

```python
#3.Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [5, 7, 6, 8, 7]
plt.scatter(x, y, color='blue', marker='o', s=100)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('My Scatter Plot')
plt.grid(True)
plt.show()
```

```python
#4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df = sns.load_dataset("iris")
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()


```
